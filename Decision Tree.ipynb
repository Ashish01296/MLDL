# Implement Decision Tree Classifier Using Scikit-Learn

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("PlayTennis.csv")
```

```python
data.head()
```

```python
X = data[['Outlook','Temperature',	'Humidity',	'Wind']]
y = data['Play Tennis']
```

```python
from sklearn import preprocessing

pre_process = preprocessing.LabelEncoder()

new_Y = pre_process.fit_transform(y)
new_X = X.apply(pre_process.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
new_X_train , new_X_test , new_Y_train ,new_Y_test = train_test_split(new_X,new_Y,train_size=0.75)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion="gini",ccp_alpha=1,splitter="best",max_depth=2)
model.fit(new_X_train,new_Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(3,3))
tree.plot_tree(model,filled=True)
```

```python
model.score(new_X_test,new_Y_test)*100
```
